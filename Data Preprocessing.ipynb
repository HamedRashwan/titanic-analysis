```python
from datetime import datetime
df['duration_years']=(datetime.now()-df['Time'])/np.timedelta64(1,'Y')
df['duration_months']=(datetime.now()-df['Time'])/np.timedelta64(1,'M')
df['duration_']=(datetime.now()-df['Time'])
df
```

```python
df['month_n']=df['Time'].dt.month
df
```

```python
def map_season(x):
    if x in [12,1,2]:
        return 'Winter'
    elif x in [3,4,5]:
        return 'Spring'
    elif x in [6,7,8]:
        return 'Summer'
    elif x in [9,10,11]:
        return 'Autumn'
    
df['season']=df['month_n'].apply(map_season)
```

```python
sns.countplot(df,x='season')
```

```python
sns.countplot(df,x='season',hue='month')
```

```python
def ext_ccprovider(x):
    x=str(x)
    if x[0]=='3':
        return 'Visa'
    elif x[0]=='4':
        return 'MasterCard'
    else:
        return 'Unknown'
    
df['cc_provider']=df['Credit Card'].apply(ext_ccprovider)
df
```
